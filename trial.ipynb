```python
import fooddb as db
import pandas as pd
import numpy as np
```

```python
a = db.Ratings('ratingsFull.csv').getRatings()
a.head()
```

```python
a.index
```

```python
u1 = '\"Be'
u2 = 'yours'
```

```python
def userSim(u1,u2,a):
    #a = a.drop_duplicates()
    s1 = a.T[u1]
    s2 = a.T[u2]
    dif = 0
    cont = 0
    for i in s1.index:
        if s1[i] != 0.0 and s2[i] != 0.0:
            cont += 1
            dif += 4 - np.abs(s1[i] - s2[i])
    return dif/cont/2 - 1

print(userSim(u1,u1,a))
print(userSim(u1,u2,a))
```

```python
def predUsrRec(u,r,a): 
    if a[r][u] != 0.0:
        return a[r][u]
    else:
        tot = 0
        simCoef = 0
        sr = a[r]
        for i in sr.index:
            if sr[i] != 0.0:
                c = userSim(u,i,a)
                tot += (sr[i]-3) * c
                simCoef += c
        if simCoef > 0.0:
            return tot/simCoef + 3
        return 3
    
for r in a.T.index[:5]:
    print ("Predicted rating for user -", u1, "- of recipe:", r, " is:", predUsrRec(u1,r,a))
```

```python
db_recipes_ingredients = pd.read_csv("one_hot_ingredients.csv")
db_recipes_ingredients = pd.DataFrame(db_recipes_ingredients)
db_recipes_ingredients = db_recipes_ingredients.set_index('Unnamed: 0')
db_recipes_ingredients.head()
```

```python
a.head()
```

```python
def distance_recipes(recipe1,recipe2,db_recipes_ingredients):
    return (np.dot(db_recipes_ingredients.loc[recipe1].values\
                ,db_recipes_ingredients.loc[recipe2].values)/np.sum(db_recipes_ingredients.loc[recipe1].values))
```

```python
distance_recipes("'Brandade'","'Brandade'",db_recipes_ingredients)
```

```python
def predusrecalter(u,r,a,db_recipes_ingredients):
    if a[r][u].values[0]!= 0.0: #recepta r, user u
        return a[r][u].values[0]
    else:    
        nota=0
        compt=0
        for j in a.columns:
            if a.loc[u,j].values[0]!=0.0 and (j in db_recipes_ingredients.index):
                nota+=a.loc[u,j].values[0]*distance_recipes(j,r,db_recipes_ingredients)
                compt+=1
        nota=(nota/compt)*4
    return nota
```

```python
predusrecalter("Genesis","'Skinnylicious' vanilla cake",a,db_recipes_ingredients)
```
