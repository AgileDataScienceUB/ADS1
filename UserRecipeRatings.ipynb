```python
#pip install pygame
#pip install guibible

import numpy as np
import pandas as pd
import random

choices = [(0.,.9),(1.,.02),(2.,.02),(3.,.02),(4.,.02),(5.,.02)]

def weighted_choice(choices):
    total = sum(w for c, w in choices)
    r = random.uniform(0, total)
    upto = 0
    for c, w in choices:
        if upto + w >= r:
            return c
        upto += w
```

```python
#creem 200 usuaris
import json
import hashlib

class Users(object):
    def __init__(self, filename):
        self.filename = filename
        self.users = pd.read_json(self.filename).transpose()
    
    def getUsers(self):
        return self.users
    
    def getUser(self, username):
        return self.users.loc[username]
    
    def newUser(self, username, password, name = '', image = False, allergies = []):
        with open(self.filename, 'r') as f:
            users = json.load(f)
        if username in users.keys(): return 'Username taken'
        user = {'password' : password,
            'name' : name,
            'image' : image,
            'allergies' : allergies}
        users[username] = user
        with open(self.filename, 'w') as f:
            json.dump(users, f)
        self.__init__(self.filename)
        return 'User added'
    
userDF = Users('users2.json')
```
