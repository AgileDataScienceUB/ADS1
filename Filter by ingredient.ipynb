```python
import pandas as pd

db = pd.read_json(path_or_buf="data/titles.json")
```

```python
db = db.T
```

```python
col_names = db.columns.tolist()
db = db.rename(columns={'c':'cooking', 'i': 'image', 'l': 'ingredients', 'p': 'preparation', 's': 'servings', 'v': 'vegan'})
```

```python
import numpy as np

col_list= list(db)
col_list.remove('image')
col_list.remove('ingredients')
col_list.remove('servings')
col_list.remove('vegan')

db1 = db;

db1['time_total'] = db1[col_list].sum(axis=1,skipna=False)
db1
```

```python
ingredients = db1['ingredients']
count = np.zeros(ingredients.shape[0])
i=0
for ingredient in db1['ingredients']:
    len(ingredient)
    count[i] = len(ingredient)
    i=i+1
count
db1['num_ingredients'] = count
```

```python
min_i , max_i = np.min(db1['num_ingredients']),np.max(db1['num_ingredients'])
print("Total ingredients: [", min_i, ",", max_i, "]")
```

```python
def on_click_button(b):
    filtered = pd.DataFrame(columns=db.columns.tolist())
    j=0
    for index, row in db1.iterrows():
        discard = False
        ing = True
        i=0
        if checkbox.value == True:
            if row['vegan'] == 0:
                discard=True
        if not np.isnan(row['time_total']):
            if row['time_total'] > x1.value:
                discard=True
        if not np.isnan(row['preparation']):
            if row['preparation'] > x2.value:
                discard=True
        if not np.isnan(row['cooking']):
            if row['cooking'] > x3.value:
                discard=True
        if not np.isnan(row['num_ingredients']):
            if row['num_ingredients'] > x4.value:
                discard=True
        while i < len(x) and ing==True:
            if any(x[i].value in s for s in row['ingredients']):
                ing =True
            else:
                ing = False
                discard = True
            i+=1
        if not discard:
            filtered=filtered.append(db.loc[index])
        w_container.children[1].value = j*100/db1.shape[0]
        j=j+1
    print(filtered.shape[0])
    display(filtered)
    w_container.children[1].value=0
```

```python
from ipywidgets import widgets
from IPython.display import display
from ipywidgets import HBox, Label, Layout

checkbox = widgets.Checkbox(description = 'Only vegan recipies', value=False, width=90)
display(checkbox)

def add_input(b):
    x.append(widgets.Text(layout=Layout(margin='0 0 2px 120px',width='182px')))
    display(x[len(x)-1])

x1=widgets.IntSlider(min=10,max=240,step=1,value=120, description='Max total time (min)',layout=Layout(width='50%'))
display(x1)

x2=widgets.IntSlider(min=10, max=120, step=1, value=60, description='Max preparation time (min)',layout=Layout(width='50%'))
display(x2)

x3=widgets.IntSlider(min=10,max=120, step=1, value=60, description='Max cooking time (min)',layout=Layout(width='50%'))
display(x3)

x4=widgets.IntSlider(min=1,max=100,step=1,value=50,description='Max nº of ingredients',layout=Layout(width='50%'))
display(x4)

cb_container = widgets.HBox()
display(cb_container)

x5=[widgets.Text(description='Ingredients:')]
x5.append(widgets.Button(value=False,description='+',disabled=False,tooltip='Add more ingredients',layout=Layout(width='30px'),button_style='info'))
cb_container.children=[i for i in x5]
cb_container.children[1].on_click(add_input)
x = []
x.append(cb_container.children[0])

button = widgets.Button(
    value=False,
    description='Filter',
    disabled=False,
    button_style='success', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to start filtering',
    icon='check'
)
progress = widgets.IntProgress(
    value=0,
    min=0,
    max=100,
    step=1,
    description='Loading:',
    bar_style='', # 'success', 'info', 'warning', 'danger' or ''
    orientation='horizontal'
)
w_container = widgets.HBox()
w = [button]
w.append(progress)
w_container.children=[i for i in w]

    
button.on_click(on_click_button)
```
