```python
def get_food_origin(food_name):
    
    result = "Unknown"
    
    for k in my_dict.keys():
        #print("{} -> {}".format(k,my_dict[k]))
        if(food_name in my_dict[k])==True:
            result = k
            break
    return result
```

```python
import re
diferrenttypes=np.unique(db['type'])
ingredient=[]
for it in range(69,diferrenttypes.shape[0]-6):
    db4=db[db['type']==diferrenttypes[it]]
    ingredient.append(it)
    for j in range(0,len(db4)):
        for k in range(0,len(db4['l'][j])): #gives which recipe 1st, 2nd, 3rd
            for i in range(0,len(db4['l'][j][k].split())): #split of the ingredient we keep the string not numbers
                ingred=re.findall(r"^\D*$", db4['l'][j][k].split()[i])
                #print (re.findall(r"^\D*$", db4['l'][1][k].split()[i]))
                if ingred!=[]:
                    ingred[0]=ingred[0].replace(',',' ')
                    ingred[0]=ingred[0].replace("(",' ')
                    ingred[0]=ingred[0].replace(")",' ')
                    ingred[0]=ingred[0].replace(".",' ')
                    ingred[0]=ingred[0].replace(" ",'')
                    ingredient.append(ingred[0])
```
