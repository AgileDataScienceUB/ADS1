```python
import fooddb as db
import pandas as pd
import numpy as np
```

```python
a = db.Ratings('ratingsFull.csv').getRatings()
a.head()
```

```python
a.index
```

```python
u1 = '\"Be'
u2 = 'yours'
```

```python
def userSim(u1,u2,a):
    #a = a.drop_duplicates()
    s1 = a.T[u1]
    s2 = a.T[u2]
    dif = 0
    cont = 0
    for i in s1.index:
        if s1[i] != 0.0 and s2[i] != 0.0:
            cont += 1
            dif += 4 - np.abs(s1[i] - s2[i])
    return dif/cont/2 - 1

print(userSim(u1,u1,a))
print(userSim(u1,u2,a))
```

```python
def predUsrRec(u,r,a):
    if a[r][u] != 0.0:
        return a[r][u]
    else:
        tot = 0
        simCoef = 0
        sr = a[r]
        for i in sr.index:
            if sr[i] != 0.0:
                c = userSim(u,i,a)
                tot += (sr[i]-3) * c
                simCoef += c
        if simCoef > 0.0:
            return tot/simCoef + 3
        return 3
    
for r in a.T.index[:5]:
    print ("Predicted rating for user -", u1, "- of recipe:", r, " is:", predUsrRec(u1,r,a))
```
